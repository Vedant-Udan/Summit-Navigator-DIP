### 1. Image Smoothing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def load_image(filename):
    """Load an image and convert it to grayscale."""
    img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    return img

def compute_histogram(image):
    """Compute the histogram of a grayscale image."""
    histogram, _ = np.histogram(image.flatten(), bins=256, range=[0, 256])
    return histogram

def moving_average_filter(histogram, kernel_size=3):
    """Apply a moving average filter to smooth the histogram."""
    kernel = np.ones(kernel_size) / kernel_size
    smoothed_histogram = np.convolve(histogram, kernel, mode='same')
    return smoothed_histogram

def plot_histograms(original, smoothed, bins):
    """Plot the original and smoothed histograms side by side."""
    plt.figure(figsize=(12, 6))

    # Plot original histogram
    plt.subplot(1, 2, 1)
    plt.bar(bins, original, color='blue', alpha=0.7, label="Original Histogram")
    plt.title("Original Histogram")
    plt.xlabel("Intensity Levels")
    plt.ylabel("Frequency")
    plt.legend()

    # Plot smoothed histogram
    plt.subplot(1, 2, 2)
    plt.bar(bins, smoothed, color='green', alpha=0.7, label="Smoothed Histogram")
    plt.title("Smoothed Histogram")
    plt.xlabel("Intensity Levels")
    plt.ylabel("Frequency")
    plt.legend()

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Enter the path to your image file
    # filename = input("Enter the image file path (e.g., 'images/sample.jpg'): ")
    filename = 'Images/23025.jpg'
    image = load_image(filename)

    # Compute histogram of the grayscale image
    histogram = compute_histogram(image)
    bins = np.arange(256)  # Intensity levels from 0 to 255

    # Smooth the histogram using a moving average filter
    smoothed_histogram = moving_average_filter(histogram)

    # Plot the original and smoothed histograms
    plot_histograms(histogram, smoothed_histogram, bins)
```
